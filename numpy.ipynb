## Funciones arrange,zeros,ones,linspace,eye

image.png

**arrange**  $(x_1,x_2,x_3)$ crea una secuencia de desde $x_1$ hasta $x_2-1$ elementos en intervalos de $x_3$

```python
array_a=np.arange(445,1590,15)
array_a
```

**zeros**  $(x_1,x_2)$ crea una matriz de $0$'s: de $x_1$ filas y $x_2$ columnas y analógo para **ones**

```python
array_z=np.zeros((5,5))
array_z
```

```python
array_o=np.ones((9,32))
array_o
```

**linspace**  $(x_1,x_2,x_3)$ crea una secuencia de desde $x_1$ hasta $x_2$ en $x_3$ elementos en intervalos de $\dfrac{x_2-x_1}{x_3-1}$

```python
array_l1=np.linspace(1,10,2)
array_l1
```

```python
array_l2=np.linspace(-10000,10000,50)
array_l2
```

**eye**$(x)$ genera la matriz indentidad de dimensión $x$

```python
array_e=np.eye(15)
array_e
```

## Numeros Aleatorios

image.png

### Numeros Aleatorios entre $0$ y $1$

**random.rand** $(m,n)$ genera una matriz de numeros aleatorios de $m$ filas y $n$ columnas

```python
array_a1 = np.random.rand(4,4)
array_a1
```

### Numeros Aleatorios de una distribucion $N (0 , 1)$

**random.randn** $(m,n)$ genera una matriz de numeros aleatorios de $m$ filas y $n$ columnas de la distribucion normal

```python
array_a2 = np.random.randn()
array_a2
```

### Numeros Aleatorios enteros

**random.randint** $(x,y,z)$ genera $z$ números aleatorios en el intervalo $[x,y]$

```python
array_a3=np.random.randint(-100,0,10)
array_a3
```

```python
array_a4=np.random.randint(900,2025,27)
array_a4
```

```python
array_a5=np.random.randint(0,5000,125)
array_a5
```

## Reshape

**reshape** $(m,n)$ cambia una lista a una matriz de $m$ filas y $columnas$

```python
matriz_r1=array_a3.reshape(5,2)
matriz_r1
```

```python
matriz_r2=array_a4.reshape(3,9)
matriz_r2
```

```python
matriz_r3=array_a5.reshape(25,5)
matriz_r3
```

## Valores Máximos y Mínimos

**max**() indica el valor máximo de un array

**argmax**() indica la posición del valor máximo de un array

**min**() indica el valor mínimo de un array

**argmin**() indica la posición del valor mínimo de un array

```python
print(matriz_r3)
print("")
print(matriz_r3.min())
print(matriz_r3.argmin())
print(matriz_r3.max())
print(matriz_r3.argmax())
```

```python
print(matriz_r1)
print("")
print("Valor mínimo ", matriz_r1.min())
print("Posición del valor mínimo", matriz_r1.argmin())
print("Valor máximo ", matriz_r1.max())
print("Posición del valor máximo", matriz_r1.argmax())
```

## Indexación en Arrays

image.png

**array** $[0]$ : Primer elemento del array

**array**$[-1]$ : Ultimo elemento del array

**array** $[x:y]$ : Array del indice $x$ al $y$

**array** $[x:]$ : Array del indice $x$ en adelante

**array** $[:y]$ : Array hasta el indice $y$

**array** $[x]=z$ : Array identico pero con el $x$ elemento $=z$

```python
array_i=np.arange(0,100,5)
array_i
```

```python
print(array_i)
print("")
print("Primer elemento",array_i[0])
print("Ultimo elemento", array_i[-1])
print("Array a partir del 10 indice", array_i[10:])
print("Array hasta el indice 6", array_i[:6])
```

```python
#Array en el indice 12 con valor 7
array_i[12]=7
array_i
```

### Indexacion Arrays de más de dos dimensiones

**array** $[0]$ : Primer fila del array

**array**$[:,0]$ : Primera columna del array

**array** $[0][0]$ : Primer fila y primera columna del array

**array**$[-1]$ : Ultima fila del array

**array** $[x:y]$ : Array de filas del indice $x$ al $y$

**array** $[x:]$ : Array de filas del indice $x$ en adelante

**array**$[:,x:]$ : Array de columnas hasta el indice $x$

**array** $[:y]$ : Array de filas hasta el indice $y$

**array**$[:,:y]$ : Array de columnas hasta el indice $y$

**array** $[x]=z$ : Array identico pero con el $x$ elemento $=z$

```python
lista_ufc=[['Islam Makhachev',27,1,0],['Jon Jones',28,1,0],['Alex Pereira',12,2,0],['Ilia Topuria',16,0,0],
 ['Merab Dvalishvili',19,4,0],['Dricus Du Plessis',23,2,0],['Belal Muhammad',24,3,0],['Tom Aspinall',15,3,0],
  ['Alexandre Pantoja',29,5,0],['Leon Edwards',22,4,0],['Alexander Volkanovski',26,4,0]]
array_ufc=np.array(lista_ufc)
```

```python
print(array_ufc)
print("")
print("Top 1",array_ufc[0])
print("Top 11", array_ufc[-1])
print("Top 5", array_ufc[:5])
print("Victorias del Top 1",array_ufc[0][1])
print("Nombres de los peleadores", array_ufc[:,0])
print("Empates de los peleadores", array_ufc[:,3])
print("Array sin empates",array_ufc[:,:3])
print("Array sin nombres",array_ufc[:,1:])
```
